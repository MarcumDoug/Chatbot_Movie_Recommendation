```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# load data
movies = pd.read_csv('../Data/IMDb movies.csv', low_memory = False)
ratings = pd.read_csv('../Data/IMDb ratings.csv', low_memory = False)
names = pd.read_csv('../Data/IMDb names.csv', low_memory = False)
principals = pd.read_csv('../Data/IMDb title_principals.csv', low_memory = False)
```

```python
# start by reviewing movie info
movies.info()
```

```python
# drop rows for with description of NaN
movies = movies.dropna(subset=['description'])

# drop columns that have no value to our recommendation
movies.drop(['metascore', 'reviews_from_users', 'reviews_from_critics'], axis=1, inplace=True)

# drop ratings, since we will be adding those back from the ratings data later
movies.drop(['avg_vote', 'votes'], axis=1, inplace=True)

# remaining null fields are for reference only, filled with string 'unknown'
movies.fillna('unknown', inplace=True)

# reset index
movies.reset_index(drop = True, inplace = True)

movies.info()
```

```python
# limit size by only listing movies available in english
movies_eng = movies.loc[movies['language'].str.contains('English')]
movies_eng = movies_eng.reset_index(drop=True)

# preprocessing of data
movies_eng = movies_eng.assign(ctry_form = [x.lower().split(", ") for x in movies_eng.country])
movies_eng = movies_eng.assign(lang_form = [x.lower().split(", ") for x in movies_eng.language])
movies_eng = movies_eng.assign(dir_form = [x.lower().split(", ") for x in movies_eng.director])
movies_eng = movies_eng.assign(writer_form = [x.lower().split(", ") for x in movies_eng.writer])
movies_eng = movies_eng.assign(act_form = [x.lower().split(", ") for x in movies_eng.actors])
movies_eng = movies_eng.assign(genre_form = [x.lower().split(", ") for x in movies_eng.genre])
movies_eng = movies_eng.assign(desc_form = [x.lower().split(", ") for x in movies_eng.description])

print(len(movies_eng))
movies_eng.head(2)
```

```python
# review names dataset
display(names.info())
names.head(3)
```

```python
# conditions for filling NaN fields
miss_reason = (pd.notnull(names.date_of_death) == True) & (pd.notnull(names.reason_of_death)== False)
miss_place = (pd.notnull(names.date_of_death) == True) & (pd.notnull(names.place_of_death) == False)      
alive = (pd.notnull(names.date_of_birth) == True) & (pd.notnull(names.date_of_death) == False)

# make copy of names data
people = names.copy()

# clean/process people data
for i in range(len(names)):
    if miss_reason[i] == True:
        people.reason_of_death[i] = 'unknown'
for i in range(len(names)):
    if miss_place[i] == True:
        people.place_of_death[i] = 'unknown'
for i in range(len(names)):    
    if alive[i] == True:
        people.place_of_death[i] = 'none - alive'
        people.death_details[i] = 'none - alive'
        people.reason_of_death[i] = 'none - alive'
        people.date_of_death[i] = 'none - alive'
for i in range(len(names)):    
    if people.spouses[i] == 0:
        people.spouses_string[i] = 'none'

people.fillna('unknown', inplace = True)
```

```python
# to avoid demographic confusion, votes will only be viewed holistically, without regard to gender, age, or location
# only keeping basic voting data
ratings_short = ratings[['imdb_title_id', 'weighted_average_vote', 'total_votes', 'mean_vote', 'median_vote']]
```

```python
# merge movies and rate_new for simple dataframe
rated_movies_org = pd.merge(movies_eng, ratings_short, on = "imdb_title_id", how = "inner") 
rated_movies_org.info()
```

```python
# due to dropping of numerous movies, a new weighted rating is formulated
R = rated_movies_org.mean_vote
v = rated_movies_org.total_votes
m = 0
C = rated_movies_org.mean_vote.mean()

rated_movies_org['new_wav'] = [round((R[i]*(v[i]/(v[i] + m))) +
                                     (C* (m/(v[i] + m))), 2) for i in range(len(rated_movies_org))]
```

```python
# remove movies with a weighted_average_vote of 4.9 or less. We will not be recommending those films. 
rated_movies = rated_movies_org[rated_movies_org.new_wav > 5].reset_index(drop = True)
rated_movies.describe()
```

```python
rated_movies.loc[rated_movies.original_title =='Iron Man']
```

```python
# create list of words for future recommendations
def list_of_words(x):
    return ' '.join(x['genre_form'])+' '+' '.join(x['dir_form'])+' '+' '.join(x['act_form'])+' '+' '.join(x['writer_form'])

# create words feature
rated_movies['words_features'] = rated_movies.apply(list_of_words, axis=1)

# display
rated_movies[['words_features']]
```

```python
# create list of words for future recommendations
def list_of_words2(x):
    return ' '.join(x['genre_form'])+' '+' '.join(x['desc_form'])

# create words feature
rated_movies['words_plots'] = rated_movies.apply(list_of_words2, axis=1)

# display
rated_movies[['words_plots']]
```
